```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread("re_snowman.jpg")
```

```python
plt.imshow(image)
```

```python
m, n, _ = image.shape
res_image = image / 255
res_image = image.reshape(m * n , 3)
```

```python
# create random cluster centroids from data istself
# we need 4 clusters (k = 4)
#-----------------------------------------------------
K = 16
randomed_idx_data =  np.random.permutation(res_image.shape[0])
centroids = res_image[randomed_idx_data[:K]]
```

```python
n = res_image.shape[1]
for t in range(11):
    
    print(f"..{t}üöÄ", end = "")
    
    # assign data to clusters using the closest distance between cetroids
    #----------------------------------------------------------------------
    AssignedClusters = np.zeros(res_image.shape[0], dtype=int)  
    for i in range(res_image.shape[0]):
        Distances = []
        for j in range(K):
            distance = np.linalg.norm(res_image[i] - centroids[j])
            Distances.append(distance)
        AssignedClusters[i] = np.argmin(Distances)


    # calculate mean datapoints for each cluster and assign the mean to the coressponding centrod using the same index
    #------------------------------------------------------------------------------------------------------------------
    centroids = np.zeros((K, n))
    for k in range(K):
        data_cluster = res_image[AssignedClusters == k]
        centroids[k] = np.mean(data_cluster, axis = 0)
    
```

```python
# convert any different points in one cluster to its centroid
#-------------------------------------------------------------
clustered_img = centroids[AssignedClusters, :]

# convert it from 3 attched vectors to 3 matrices
#-------------------------------------------------
clustered_img = np.reshape(clustered_img, image.shape)

# Happy end
#-----------
plt.imshow(clustered_img / 255)
```
